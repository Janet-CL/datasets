## DOHMH NYC Restaurant Inspection Dataset

### Importing libraries and modules

```python
import pandas as pd
import numpy as np
```

### Loading Original Dataset

```python
raw_data = pd.read_csv('DOHMH_New_York_City_Restaurant_Inspection_Results_20231025.csv')
```

```python
raw_data.head(10)
```

```python
raw_data.info()
```

```python
raw_data.columns
```

```python
df = raw_data.copy()
```

```python
df.head()
```

```python
drop_labels = ['DBA', 'BUILDING', 'STREET', 'ACTION', 'VIOLATION DESCRIPTION', 'Community Board', 
               'Council District', 'Census Tract', 'BIN', 'BBL', 'NTA', 'Location Point1']
```

```python
df = df.drop(labels=drop_labels, axis=1)
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.dtypes
```

```python
df = df.drop('PHONE', axis=1)
```

```python
df.isnull().sum()
```

```python
df[['ZIPCODE', 'SCORE']] = df[['ZIPCODE', 'SCORE']].fillna(0)
```

```python
df[['ZIPCODE', 'SCORE']] = df[['ZIPCODE', 'SCORE']].astype(int)
```

```python
df
```

```python
df.iloc[:,[4,9,10]]
```

```python
df['INSPECTION DATE'] = pd.to_datetime(df['INSPECTION DATE'], format='%m/%d/%Y')
```

```python
df['GRADE DATE'] = pd.to_datetime(df['GRADE DATE'], format='%m/%d/%Y')
```

```python
df['RECORD DATE'] = pd.to_datetime(df['RECORD DATE'], format='%m/%d/%Y')
```
